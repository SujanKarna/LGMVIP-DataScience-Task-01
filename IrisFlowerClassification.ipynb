# Iris Flowers Classification

### Import necessary python libraries

```python
import pandas as pd
##import train test split from sklearn
from sklearn.model_selection import train_test_split
##import seaborn -> Seaborn is a library for making statistical graphics in Python
import seaborn as sns
```

### Import Iris dataset

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
X = iris.data.features 
y = iris.data.targets 
  
# metadata 
print(iris.metadata) 
  
# variable information 
print(iris.variables) 
```

### Divide the dataset into train dataset and test dataset using train_test_split from sklearn

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)
```

### Using decision tree classifier from scikit learn

```python
## import DecisionTreeClassifier from sklearn
from sklearn.tree import DecisionTreeClassifier
## import tree from sklearn to plot the decision tree
from sklearn import tree
```

```python
## entropy is selected as attribute selection measure
clf = tree.DecisionTreeClassifier(criterion="entropy")
##fit the model with our training dataset
clf = clf.fit(X_train, y_train)
```

```python
## import plt from matplotlib to define plot size
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
tree.plot_tree(clf,filled = True)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
##import accuracy_score to measure how accurate our model is
from sklearn.metrics import accuracy_score, classification_report
```

```python
##print the score using accuracy_score
score = accuracy_score(y_pred , y_test)
print(score)
```

```python
##use classification_report for more detailed report whict provides different scores including f1 score.
print(classification_report(y_pred, y_test))
```

```python
## Confusion Matrix Display
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
